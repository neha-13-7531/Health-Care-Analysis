```python
#  univariate for categorical column
def univariate_categorical(categorical_columns, df):
    for column in categorical_columns:
        plt.figure(figsize=(8,6))
        df[column].value_counts().head().plot(kind='bar')
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.show()
categorical_columns = ['Gender', 'Diagnosis', 'Outcome', 'Readmission']
univariate_categorical(categorical_columns, df)
```

**1.Distribution of Gender**
- The highest gender count is **other** is above 1600.
- The count for **female** and **male** gender category is roughly the same.

**2.Distribution of Diagnosis**
- The graph indicates that hypertension has the highest count, followed closely by diabetes and heart disease, which are roughly equal. Flu and COVID-19 have lower counts.

**3.Distribution of Outcome**
- The outcomes "Recovered," "Improved," "Deteriorated," and "Deceased" all have approximately equal counts, around 1200.

**4.Distribution of Readmission**
- The highest count is Readmitted patient.

```python
# Bivariate Analysis for categorical column

cat_col=['Gender','Outcome','Diagnosis']
def bivariate_analysis(n):
    plt.figure(figsize=(14,10))
    for i,col in enumerate(n,1):
        plt.subplot(2,2,i)
        plt.title(f"count of {col}")
        sns.countplot(x=df[col],hue='Readmission',data=df)
    plt.show()
bivariate_analysis(cat_col)
```

**1.Count of Gender**
- The "Other" gender category has a higher count of readmitted patients (around 850) compared to those not readmitted (around 800), indicating a slightly higher likelihood of readmission.
- Males have a relatively balanced count between readmitted (around 800) and not readmitted (around 800) patients, suggesting that readmission rates are roughly equal for this gender.
- Females have a lower count of readmitted patients (around 750) compared to those not readmitted (around 800), indicating a slightly lower likelihood of readmission.

**2.Count of Outcome**
- In the "Deceased" category, the count for "Yes" readmission is slightly higher than for "No" readmission.
- The "Deteriorated" category has the highest count for both "Yes" and "No" readmission, suggesting that patients who deteriorate are more likely to be readmitted.
- The identical counts for "Yes" and "No" readmission in the "Improved" and "Recovered" categories imply that readmission status may not be a significant factor in determining the outcome for patients who improve or recover.

**3.Count of Diagnosis**
- In Heart Diesease the count of readmitted patient is greater than non-admitted patient.
- In Hypertension diagnosis count of non-admitted patient is greater than admitted patient.
- In Diabetes count of non-admitted patient is greater than admitted patient.
- In Flu column non-admitted patient is greater than admitted patient.
- In COVID-19 readmitted patient is greater than non-admitted patient.
